```python
#bag of words
import numpy as np
import pandas as pd
temp_df=pd.read_csv('IMDB Dataset.csv')
temp_df.head()
```

```python
df.drop_duplicates(inplace=True) #remove duplcate review
```

```python
df.duplicated().sum()
```

```python
import re
def remove_tags(raw_text):#remove html tages
    cleaned_text=re.sub(re.compile('<.*?>'),'',raw_text)
    return cleaned_text
```

```python
df['review']=df['review'].apply(remove_tags)
```

```python
df
```

```python
df['review']=df['review'].apply(lambda x:x.lower())#convert in lower case
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords#remove stopwords
sw_list=stopwords.words('english')
df['review']=df['review'].apply(lambda x:[item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

```python
x=df.iloc[:,0:1]
y=df['sentiment']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
y=en.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
x_train.shape
```

```python
#applying bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_train_bow=cv.fit_transform(x_train['review']).toarray()
x_test_bow=cv.transform(x_test['review']).toarray()
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train_bow,y_train)
```

```python
y_pred=model.predict(x_test_bow)
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
#
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train_bow,y_train)
y_pred=rf.predict(x_test_bow)
accuracy_score(y_test,y_pred)
```

```python
import nltk
nltk.download('punkt_tab') # Download the punkt_tab resource
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec,KeyedVectors
import gensim

story=[]
for dr in df['review']:
    raw_sent=sent_tokenize(dr)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
model=gensim.models.Word2Vec(window=10,min_count=2,workers=4)
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
len(model.wv.index_to_key)
```

```python
def document_vector(dr):
  dr=[word for word in dr.split() if word in model.wv.index_to_key]
  if len(dr)>0:
    return np.mean(model.wv[dr],axis=0)
```

```python
document_vector(df['review'][0])
```

```python
from tqdm import tqdm
x=[]
for dr in tqdm(df['review']):
  x.append(document_vector(dr))
```

```python
x=np.array(x)
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
y=en.fit_transform(df['sentiment'])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
x_train.shape
```

```python
# Instead of:
# from sklearn.native_bayes import GaussianNB

# Use:
from sklearn.naive_bayes import GaussianNB  # Correct import path for GaussianNB
model = GaussianNB()  # Initialize a new GaussianNB model (optional, if model was intended to be re-used, this is not necessary)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
#confusion_matrix(y_test,y_pred)
```
