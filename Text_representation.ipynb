```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
bow=cv.fit_transform(df['text'])
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(['campusx watch and write comment of  ']).toarray()
```

```python
#disadvantages of bag of words
#ordering issue in words,changes in statement not capture
```

```python
#techniques=Bag of n-grams
#Bi-grams,uni-grams,tri-grams
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,3))
bow=cv.fit_transform(df['text'])
```

```python
#vocab
print(len(cv.vocabulary_))
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
#advantages & disadvantages
#dimesionality increases,time complexcity increases
```

```python
#techniques=Tf-Idf(term frequency-inverse doc.frequency)
#tf-particular term kitna important hai document
#idf=corpus ke liye kitna important hai
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
